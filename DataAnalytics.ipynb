```python
import pandas as pd

df = pd.read_csv("automobile2.csv")
df
```

```python
df.columns
```

```python
import numpy as np

df.replace("?", np.NaN,inplace = True)
df
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
X
```

```python
missing_data = df.isnull()
print(missing_data.columns.values.tolist())
print()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
#We replaced all the float/int missing values with the mean

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean") 
imputer.fit(X[:,0:15])

#Saving the transformed X features var in the same matrix
X[:,0:15] = imputer.transform(X[:,0:15])
```
